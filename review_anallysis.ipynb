# 🛠️📝 Text Preprocessing and Visualization 📝🛠️

### 🔍📜 Text Preprocessing 📜🔍

- **Stopwords Handling**:
  - **Loading English stopwords** and removing 'not' to preserve negation in sentiment analysis.

- **Stemming**:
  - **Initializing Porter Stemmer** for reducing words to their root forms.

- **Text Cleaning Function**:
  - **Removing non-alphanumeric characters** and stemming words while excluding stopwords.

- **Applying Preprocessing**:
  - **Cleaning the 'Review' column** by applying the preprocessing function to each review.

  - **Sample Cleaned Review**:
    - Review 50: 'we order duck rare pink tender insid nice char outsid'

### 🌟✨ Word Cloud Generation and Visualization ✨🌟

- **Combining Reviews by Sentiment**:
  - **Positive Reviews**: Creating a single string containing all positive reviews.
  - **Negative Reviews**: Creating a single string containing all negative reviews.

  - **Positive Reviews Word Cloud**:
    - **Creating**: Using `WordCloud` to generate a word cloud for positive reviews.

### 🔢📜 Combining All Reviews 📜🔢

- **Combining All Reviews**:
  - **List of all reviews** converted to text.
  - **Total Reviews**: 900 reviews in the dataset.


```python
stop =stopwords.words('english')
if 'not' in stop:
    stop.remove('not')
pt=PorterStemmer()

def preprocessing(text):
    clean_txt=[]
    tex=r.sub('[^a-zA-Z0-9]',' ',text)
    for i in tex.split():
        if i not in stop:
            stemm=pt.stem(i.lower())
            clean_txt.append(stemm)
    return  " ".join(clean_txt)  

    
    
```

```python
df['Review']=df['Review'].apply(preprocessing)
```

# 🔠🧠 Text Vectorization and Data Splitting 🧠🔠

### ✨🔤 Text Vectorization 🔤✨

- **TF-IDF Vectorization**:
  - **Converting text data** into numerical format using `TfidfVectorizer`.
  - **Fit and transform** the entire text corpus into a TF-IDF matrix.

### 🔍🔢 Preparing Features and Labels 🔢🔍

- **Feature Matrix and Labels**:
  - **X**: TF-IDF feature matrix.
  - **y**: Labels (Liked or Not Liked).

### 🔄🧩 Train-Test Split 🧩🔄

- **Splitting the data** into training and testing sets using `train_test_split`.
  - **Training Set**: 80% of the data.
  - **Testing Set**: 20% of the data.
  - **Random State**: Ensuring reproducibility with `random_state=42`.

### 📊🔢 Data Shapes 🔢📊

- **Shape of Training and Testing Data**:
  - **X_train Shape**: (720, 1541)
  - **X_test Shape**: (180, 1541)
  - **y_train Shape**: (720,)
